```python
import requests
from bs4 import BeautifulSoup
import time

# Define the base URL of the forum and the initial page number
base_url = 'https://community.fitbit.com/t5/Sleep-Well/bd-p/sleep'
page_number = 1

# Initialize an empty list to store thread URLs
all_thread_urls = set()

# Create a session to manage cookies
session = requests.Session()

while page_number <= 41:  # Adjust the maximum number of pages as needed
    # Construct the URL for the current page of the forum
    forum_url = f'{base_url}/page/{page_number}'
    
    try:
        # Something to know
        print("Scaning page " + str(page_number))
        # Send an HTTP GET request to the forum page
        r_forum = session.get(forum_url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'})
        
        # Check if the request was successful (status code 200)
        r_forum.raise_for_status()
        
        # Parse the HTML content with BeautifulSoup
        soup_forum = BeautifulSoup(r_forum.content, 'html.parser')
        
        # Find all thread links on the page
        thread_links = soup_forum.find_all('a', class_='page-link lia-link-navigation lia-custom-event')
        
        # Extract the href attributes and concatenate them into the list
        thread_urls = [f'https://community.fitbit.com{link.get("href")}' for link in thread_links]
        
        # Extend the list of thread URLs with the URLs from this page
        all_thread_urls.update(thread_urls)
        time.sleep(2)  # Add a delay to avoid overloading the server
        # Increment the page number for the next iteration
        page_number += 1  
        
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        break
```

```python
def scrape_fitbit_forum_comments(base_url):
    # Initialize variables
    page_number = 1
    unique_comments = set()  # Store unique comments

    while True:
        # Construct the URL for the current page of the thread
        url_beitrag = f'{base_url}/page/{page_number}'
        
        # Send an HTTP GET request to the URL
        r_beitrag = requests.get(url_beitrag)
        
        # Check if the request was successful (status code 200)
        if r_beitrag.status_code == 200:
            # Parse the HTML content with BeautifulSoup
            soup_beitrag = BeautifulSoup(r_beitrag.content, 'html.parser')
            
            # Find all comment elements
            comment_elements = soup_beitrag.find_all('div', class_='lia-message-body-content')
            
            if not comment_elements:
                # No more comments found on this page; terminate the loop
                break
            
            for comment_element in comment_elements:
                # Extract text content from the comment element
                comment_text = comment_element.get_text().strip()
                
                # Check if the comment is unique (not already in the set)
                if comment_text not in unique_comments:
                    # Add the unique comment to the set
                    unique_comments.add(comment_text)
            
            # Increment the page number for the next iteration
            page_number += 1
            
            # Rest for a moment to avoid overloading the server
            time.sleep(1)

            # Check if there's a "Next" button on the page; if not, terminate the loop
            next_button = soup_beitrag.find('span', class_='lia-paging-page-next')
            if not next_button:
                break
        else:
            # Break the loop if the request fails (reached the end of pages)
            break
    
    return unique_comments
```

```python
all_comments = set()
indx = 1
for url in all_thread_urls:
    new_comments = scrape_fitbit_forum_comments(url)
    all_comments.update(new_comments)
    indx += 1
    if indx in set(range(1,1000,10)):
        print('processing url ' + str(indx))
```
