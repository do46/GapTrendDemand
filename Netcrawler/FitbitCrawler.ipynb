```python
import requests
from bs4 import BeautifulSoup
import time  # Import the time module

def extract_beitrag_text(url):
    r_beitrag = requests.get(url)
    soup_beitrag = BeautifulSoup(r_beitrag.content, 'html.parser')
    # Extract text from all div elements with class="lia-message-body-content"
    beitrag = [div.get_text() for div in soup_beitrag.find_all('div', class_='lia-message-body-content')]
    # Join the text into a single string
    beitrag_corpus = '\n'.join(beitrag)
    return beitrag_corpus

# Base URL for Fitbit community pages
base_url = 'https://community.fitbit.com/t5/Sleep-Well/bd-p/sleep/page/'

# Number of pages to scrape
num_pages = 3  # Adjust this as needed

# Sleep time in seconds between requests
sleep_time = 2  # Adjust this as needed

# List to store the extracted text from all pages
extracted_texts = []

# Loop through each page and extract links
for page_num in range(1, num_pages + 1):
    page_url = f'{base_url}{page_num}'
    page_r = requests.get(page_url)
    page_soup = BeautifulSoup(page_r.content, 'html.parser')
    
    # Find all matching <a> elements on the current page
    a_elements = page_soup.find_all('a', class_='page-link lia-link-navigation lia-custom-event')
    
    # Extract the href attributes and concatenate them into a list
    href_list = [f'https://community.fitbit.com{a.get("href")}' for a in a_elements]
    
    # Apply the function to each URL in the list and store the results
    for url in href_list:
        beitrag_text = extract_beitrag_text(url)
        extracted_texts.append(beitrag_text)
        
        # Add a sleep time between requests
        time.sleep(sleep_time)

# Now, extracted_texts contains the extracted text from all pages and discussions with delays between requests
```
