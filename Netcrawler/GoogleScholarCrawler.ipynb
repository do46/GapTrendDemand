```python
import requests
from bs4 import BeautifulSoup
import time
import re
import pandas as pd
from datetime import datetime, timedelta

def Crawler_GoogleScholar_SummaryOnly(schlagwörterlist):
    booklist = []
    for wörter in schlagwörterlist:
        for page in range(0,3):
            """
            Loop through all pages
            """
            #age=[]
            url = f'https://scholar.google.com/scholar?start={page}0&q={wörter}&hl=de&scisbd=1&as_sdt=0,5'
            r = requests.get(url)
            #print(r)
            #r have to be 200. 403: no go
            soup = BeautifulSoup(r.content, 'html.parser')
            #print(soup)
            #copy "soup" to Word to analyze :)
            articles = soup.find_all('div', class_='gs_r gs_or gs_scl')
            if not articles:
                break
            #page is empty, break loop
            try:
                age = re.findall(r'\d+', soup.find('span', class_='gs_age').text)
            except AttributeError:
                #print('age')
                pass
            #page without datum

            if int(age[0]) > 1500:
                break
            #page is older than 30 days, stop loop

            for article in articles:
                #loop through all article in 1 page
                """
                ignore Error when there are no paper's "age"
                """

                try:
                    age = re.findall(r'\d+', article.find('span', class_='gs_age').text)
                except AttributeError:
                    age[0]=0

                erscheinung=datetime.today().date()- timedelta(days=int(age[0]))

                if int(age[0]) <= 1500:
                    #only save books/article if they are published within the last 30 days
                    title = article.find('h3', class_='gs_rt').text
                    title = title.replace('[HTML]','')
                    autor = article.find('div', class_='gs_a').text.split('-')[0].strip()
                    try:
                        summary = article.find('div', class_='gs_rs').text.split('-', 1)[1].strip()
                    except IndexError:
                        summary = "Keine Zusammenfassung"
                    booklink = article.find('a')['href']

                    book = {
                        'date': datetime.today().date(),
                        'erscheinung': str(erscheinung),
                        'title': title,
                        'autor': autor,
                        'summary': summary,
                        'booklink': booklink
                    }
                    booklist.append(book)
                else:
                    break
            print(f"Processing Google Scholar SummaryOnly finding {wörter} in page {page}...")
            #some text to pass time
            time.sleep(5)
            #time sleep so the page will not be overload by our request
    df = pd.DataFrame(booklist)
    return df

def Crawler_GoogleScholar_Alltext(schlagwörterlist):
    booklist = []
    for wörter in schlagwörterlist:
        for page in range(0,3):
            """
            Loop through all pages
            """
            #age=[]
            url = f'https://scholar.google.com/scholar?start={page}0&q={wörter}&hl=de&scisbd=2&as_sdt=0,5'
            r = requests.get(url)
            #print(r)
            #r have to be 200. 403: no go
            soup = BeautifulSoup(r.content, 'html.parser')
            #print(soup)
            #copy "soup" to Word to analyze :)
            articles = soup.find_all('div', class_='gs_r gs_or gs_scl')
            if not articles:
                break
            #page is empty, break loop
            try:
                age = re.findall(r'\d+', soup.find('span', class_='gs_age').text)
            except AttributeError:
                #print('age')
                pass
            #page without datum

            if int(age[0]) > 1500:
                break
            #page is older than 30 days, stop loop

            for article in articles:
                #loop through all article in 1 page
                """
                ignore Error when there are no paper's "age"
                """

                try:
                    age = re.findall(r'\d+', article.find('span', class_='gs_age').text)
                except AttributeError:
                    age[0]=0

                erscheinung=datetime.today().date()- timedelta(days=int(age[0]))

                if int(age[0]) <= 1500:
                    #only save books/article if they are published within the last 30 days
                    title = article.find('h3', class_='gs_rt').text
                    title = title.replace('[HTML]','')
                    autor = article.find('div', class_='gs_a').text.split('-')[0].strip()
                    try:
                        summary = article.find('div', class_='gs_rs').text.split('-', 1)[1].strip()
                    except IndexError:
                        summary = "Keine Zusammenfassung"
                    booklink = article.find('a')['href']

                    book = {
                        'date': datetime.today().date(),
                        'erscheinung': str(erscheinung),
                        'title': title,
                        'autor': autor,
                        'summary': summary,
                        'booklink': booklink
                    }
                    booklist.append(book)
                else:
                    break
            print(f"Processing Google Scholar Alltext finding {wörter} in page {page}...")
            #some text to pass time
            time.sleep(5)
            #time sleep so the page will not be overload by our request
    df = pd.DataFrame(booklist)
    return df
```
